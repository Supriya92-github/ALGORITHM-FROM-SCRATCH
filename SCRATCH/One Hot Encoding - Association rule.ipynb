```python
#import libraries
import pandas as pd
import numpy as np
import os
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#change directory
os.chdir("C:\\Users\\92sup\\Downloads\\PYTHON") 
#read data
fm=pd.read_csv("lastfm.csv")
```

```python
#sample the data
fm=fm.head(500)  #limit the data set
```

```python
#form the transactions as per user
records=fm.groupby(['user'])['artist'].apply(list).values.tolist()
#print the transactional data
records
```

```python
#keep the requisite columns only
fm = fm[['user','artist']]
fm
```

```python
#find number of unique artists and users
artists = set(fm.artist)  
users = set(fm.user) 
print(len(artists),len(users))
```

```python
#create a dataframe with unique artists as column names
df = pd.DataFrame(columns = artists)
df
```

```python
#loop to fill the values in dataframe in binary format
for user in set(fm.user):
    for i in records:
        temp={}
        for  element in i:
            temp[element]=1
        df = df.append(temp, ignore_index = True)
df
```

```python
#set the null values as 0
df[df.isnull()] = 0 
df
```

```python
#apply the apriory rule with min support hence to create the frequently used transaction pair
freq_itemset = apriori(df=df, min_support = 0.1, use_colnames = True) # Finding the support for the artists
freq_itemset
```

```python
#apply association rule
rules = association_rules(freq_itemset, min_threshold = 0.5)
rules
```
