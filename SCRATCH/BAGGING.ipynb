# BAGGING

```python
pred={}

#Number of bagging rounds
n=20

data = iris

for j in range(0,n):
    #Draw a bootstrap sample
    boot= iris.sample(len(iris), replace = True)
    uniq_index=np.unique(boot.index)
    test_index=[]

    for i in range(0,len(iris)):
        if i not in uniq_index:
            test_index.append(i)
        
    #out of bag data
    test=iris.iloc[test_index]
    
    #SPLITTING X & Y
    x_boot=boot.drop("species", axis=1)
    y_boot=boot[["species"]]

    x_test=test.drop("species", axis=1)
    y_test=test[["species"]]
    
    #Training the model on boot strap sample
    tree= DecisionTreeClassifier(criterion="gini", min_samples_leaf=5)
    tree.fit(x_boot,y_boot)

    #OOB PREDICTION
    y_pred=tree.predict(x_test)

    #attaching the pred to the prediction data frame
    pred1 = np.repeat(np.nan,len(iris))
    pred1=np.array(pred1,dtype='<U10')
    pred1[test_index]=y_pred
    key = "Pred" + str(j+1)
    pred[key] = pred1
    
```

```python
pred=pd.DataFrame(pred)
pred
```

```python
# here to find the frequency of each class

def freq(x):
    d={}
    for i in x:
        d[i]=d.get(i,0) + 1
        
    # Remove the nan key
    d.pop('nan',None)
    key = list(d.keys())
    val=list(d.values())
    return(key[val.index(max(val))])
```

```python
row1=pred.iloc[0,]
freq(row1)
```

```python
#obtaining the final prediction for all the predicted observations

prediction=[]

for index in range(0,len(pred)):
    prediction.append(freq(pred.iloc[index,]))
    
    
prediction
```

```python
# Attach the "predicted" column to the data which we have got as output by aggregating the outputs of all the models 

data=iris.copy()
data["predicted"]=prediction
data
```

```python
conf_matrix= confusion_matrix(data.species,data.predicted)
conf_matrix
```
